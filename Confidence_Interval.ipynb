```python
from scipy import stats
import numpy as np
import pandas as pd
```

```python
glaxo_df = pd.read_csv('glaxo_df.csv')
glaxo_df.columns = glaxo_df.columns.str.strip()
glaxo_df
```

```python
glaxo_df_ci = stats.norm.interval(0.99,
loc = glaxo_df.gain.mean(),
scale = glaxo_df.gain.std())

print( 'Mean at 95% confidence interval is:', np.round(glaxo_df_ci, 4))
```

```python
glaxo_df.shape[0]-1
```

```python
glaxo_df_ci = stats.t.interval(0.95,
loc = glaxo_df.gain.mean(),
scale = glaxo_df.gain.std(), df=1725)

print( 'Mean at 95% confidence interval is:', np.round(glaxo_df_ci, 4))
```
