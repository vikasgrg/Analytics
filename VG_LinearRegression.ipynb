**Linear Regression Example**

---


Equation of the Best-Fit Line
y=mx+b

y is the predicted value (dependent variable)
x is the input (independent variable)
m is the slope/Gradient/coefficient of the line (how much y changes when x changes) if the slope is 5, it means that for every 1-unit increase in x, the value of y increases by 5 units.
b is the intercept (the value of y when x = 0)

Residual=yᵢ− y^ᵢ
= actual value - predeicted value

Sumofsquarederrors(SSE)=Σ(yᵢ− y^​ᵢ)²

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
url = "https://raw.githubusercontent.com/vikasgrg/KagleDS/refs/heads/main/Student_Marks.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()
# df = df.iloc[:,1:3]
df.describe()
```

```python
# model = LinearRegression()
# model.fit(X_train, y_train)
plt.xlabel('Time_Studied')
plt.ylabel('Marks')
plt.scatter(df['time_study'], df['Marks'], color='red', marker='+')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(df[['time_study']], df.Marks)
```

```python
df.shape
```

```python
# Correct prediction for time_study = 7
# X = np.array(7.811).reshape(-1, 1)  # Reshape to (1, 1)
reg.predict(df[['time_study']])
# df[['time_study']].iloc[0,0]
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# y = mx + b
y = reg.coef_ * 7.811 + reg.intercept_
y
```

```python
# model = LinearRegression()
# model.fit(X_train, y_train)
plt.xlabel('Time_Studied')
plt.ylabel('Marks')
plt.scatter(df['time_study'], df['Marks'], color='red', marker='+')
plt.plot(df['time_study'], reg.predict(df[['time_study']]), color='blue')
plt.show()
```
